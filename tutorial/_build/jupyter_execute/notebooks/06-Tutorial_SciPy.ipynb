## Exemplo: integração numérica


Dada a equação diferencial ordinária


$$\frac{dy(t)}{dt} = -y(t) + 1$$
$$y(0) = 0$$


Encontre uma solução numérica para a equação diferencial com condição inicial associada. Expanda o horizonte temporal até encontrar uma solução estável.

```python
%matplotlib widget
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(y,t):
    dydt = -y + 1.0
    return dydt

# initial condition
y0 = 0

# time points
t = np.linspace(0,5)

# solve ODE
y = odeint(model,y0,t)

# plot results
fig = plt.figure()
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```
